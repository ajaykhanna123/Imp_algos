```python
df_dates=pd.read_csv('country_date_formats.csv')
```

```python
import re
import random
```

```python
dates=[]
fill_years_4_digits=[str(i) for i in range(1900,2026)]
fill_years_2_digits=[str(i) for i in range(19,41)]
fill_months=[str(i) for i in range(1,13)]
fill_days=[str(i) for i in range(1,29)]
for date_format in df_dates['Date Format'].values:
    date=date_format
    if re.search('yyyy',date_format,re.IGNORECASE):
        date=re.sub('yyyy',random.choice(fill_years_4_digits),date,flags=re.IGNORECASE)
    elif re.search('yy',date_format,re.IGNORECASE):
        date=re.sub('yy',random.choice(fill_years_2_digits),date,flags=re.IGNORECASE)
    
    if re.search('MM',date_format,re.IGNORECASE) or re.search('M',date_format,re.IGNORECASE):
        date=re.sub('yy|y',random.choice(fill_months),date,flags=re.IGNORECASE)
    if re.search('MM',date_format,re.IGNORECASE) or re.search('M',date_format,re.IGNORECASE):
        date=re.sub('MM|M',random.choice(fill_months),date,flags=re.IGNORECASE)
    if re.search('dd',date_format,re.IGNORECASE) or re.search('d',date_format,re.IGNORECASE):
        date=re.sub('dd|d',random.choice(fill_days),date,flags=re.IGNORECASE)
    dates.append(date)
        
```

```python
df_patients=pd.read_csv('patients.csv')
```

```python
df_patients.head()
```

```python
addresses=df_patients['ADDRESS'].values
cities=df_patients['CITY'].values
names=[]
for name in  df_patients['FIRST']+ " " +df_patients['LAST']:
    names.append(re.sub("\d+","",name))
```

```python
zipcodes=df_patients['ZIP'].dropna(axis=0).astype(int).astype(str).values
```

```python
df_patients.columns
```

```python
# passports

url='https://raw.githubusercontent.com/vaasha/Data-Analysis_in-Examples/master/EDA_Passport_Numbers/data.csv'
passports=pd.read_csv(url)['PassportNumber'].dropna(axis=0).values
```

```python
# email ids
file='email_domains.txt'
f=open(file,'r')
domains=f.read().split("\n")
names=df_patients['LAST'].dropna(axis=0).values
emails=[]
for name in names:
    emails.append(name+"@"+random.choice(domains))
```

```python
# phone numbers
file='phone_no.txt'
f=open(file,'r')
phone_data=f.read().split("\n")

contacts=[]
for numbers in phone_data:
    contacts.append(numbers.split(",")[1].strip())
```

```python
# medical records
import string
import random

N = 7
medical_records=[]
for i in range(1000):
    random_name= ''.join(random.choices(string.digits, k = N))
    medical_records.append(random_name)
    

medical_prefix=["MR","ID","CA","DS"]
medical_ids=[]

for i in medical_records:
    medical_ids.append("".join(random.choices(medical_prefix))+str(i))
```

```python
# account numbers

account_numbers=[]
N = 10
for i in range(1000):
    random_name= ''.join(random.choices(string.digits, k = N))
    account_numbers.append(random_name)
```

```python
# urls
file='urls.txt'
f=open(file,'r')
urls_data=f.read().split("\n")

urls=[]

for i in urls_data:
    urls.append(re.sub("'|,","",i))
```

```python
# IP address
df_ips=pd.read_csv('ipv4.csv')
df_ips.columns=['ip1','ip2','states']
ip_address=df_ips['ip1'].values
```

```python
diseases=['cancer','heart attack','fever','cold','cough','Diarrhea','headache','stomach ache']
```

```python
sample_sentences=["""Mr <Name> is suffering from <Disease> . Patient SSN <SSN> . He arrived on <Date> . Patient is from <Address> <City> , <Zipcode> .
Advised him to reach <URL> . Patient account number <Account> . Medical record Id <Medical> . Patient <Contact> Email is <Email>
Patient Id <Passport> . IP address <IP>""","""Mr <Name> has  <Disease> . Patient SSN <SSN> . He came on <Date> . Patient home address is  <Address> <City> , <Zipcode> .
we gave him <URL> . Patient account number <Account> . Medical record Id <Medical> . Patient <Contact> Email is <Email>
Patient Id <Passport> . Here is the IP <IP>"""]
```

```python

final_list_data=[]
for _ in range(10000):
    name=random.choice(names)
    disease=random.choice(diseases)
    ssn=random.choice(snns)
    date=random.choice(dates)
    address=random.choice(addresses)
    city=random.choice(cities)
    zipcode=random.choice(zipcodes)
    url=random.choice(urls)
    acc_no=random.choice(account_numbers)
    med_id=random.choice(medical_ids)
    contact=random.choice(contacts)
    email=random.choice(emails)
    passport=random.choice(passports)
    ip=random.choice(ip_address)
    rep={"<Name>":name,"<Disease>":disease,"<SSN>":ssn,"<Date>":date,
    '<Address>':address,'<City>':city,"<Zipcode>":zipcode,
    "<URL>":url,"<Account>":acc_no,"<Medical>":med_id,
    "<Contact>":contact,"<Email>":email,"<Passport>":passport,
    "<IP>":ip}
    
    pattern = re.compile("|".join(rep.keys()))
    text = pattern.sub(lambda m: rep[re.escape(m.group(0))], random.choice(sample_sentences))
    
    final_list_data.append([text,name,disease,ssn,date,address,city,zipcode,url,acc_no,med_id,contact,email,passport,ip])
    
```

```python
final_data=pd.DataFrame(columns=['Sentence','Name','Disease','SSNs','Dates','Address','City','Zipcode','Url',"Account","Medical Record","Contact","Email","Passport","IP Address"],data=final_list_data)

    
```
